# Feature selection and Feature Engineering

```python
model_data = pd.get_dummies(data_set)
```

I dropped some labels to avoid the dummy variable trap.

```python
model_data = model_data.drop(['Y_0', 'X2_female', 'X3_0', 'X3_others', 'X4_0','X4_married', 'X3_high school'], axis = 1)
```

# Splitting the dataset

```python
X = model_data.iloc[:, :-1].values
Y = model_data.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Training the model

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

# Model Evaluation

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import precision_score, recall_score

print('Precisions:', precision_score(y_test, y_pred))
print('Recall:', recall_score(y_test, y_pred))


from sklearn.metrics import f1_score

print('f1 score:', f1_score(y_test, y_pred))
```
